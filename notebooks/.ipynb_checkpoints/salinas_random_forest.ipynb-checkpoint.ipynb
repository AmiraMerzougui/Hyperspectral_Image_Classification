```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
salinas = loadmat('../data/salinas/Salinas_corrected.mat')
salinas_gt = loadmat('../data/salinas/Salinas_gt.mat')
```

```python
print(salinas.keys())
print(salinas_gt.keys())
```

```python
X = salinas['salinas_corrected']
y = salinas_gt['salinas_gt']

print("Hyperspectral cube shape:", X.shape)
print("Ground truth shape:", y.shape)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(X[:, :, 9], cmap='gray')
plt.title('Salinas  Band 10')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.imshow(y, cmap='tab20')
plt.title('Salinas Ground Truth')
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
unique, counts = np.unique(y, return_counts=True)

for u, c in zip(unique, counts):
    print(f"Class {u}: {c} pixels")
```

```python
mask = y > 0
print("Number of labeled pixels:", np.sum(mask))
```

```python
X_pixels = X[mask]
y_pixels = y[mask]

print("X_pixels shape:", X_pixels.shape)
print("y_pixels shape:", y_pixels.shape)
```

```python
print("Min label:", y_pixels.min())
print("Max label:", y_pixels.max())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_pixels_scaled = scaler.fit_transform(X_pixels)

print("Scaled data shape:", X_pixels_scaled.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_pixels_scaled,
    y_pixels,
    test_size=0.3,
    random_state=42,
    stratify=y_pixels
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
print("Random Forest trained successfully!")
```

```python
from sklearn.metrics import accuracy_score

y_pred = rf.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Test Accuracy:", acc)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot(cmap='viridis')
plt.title("Random Forest  Confusion Matrix")
plt.show()
```

```python
y_all_pred = rf.predict(X_pixels_scaled)
print("All labeled pixels predicted!")
```

```python
# Create empty map with same shape as ground truth
prediction_map = np.zeros(y.shape, dtype=np.int32)

# Put predictions back into image shape
prediction_map[mask] = y_all_pred

print("Prediction map shape:", prediction_map.shape)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(prediction_map, cmap='tab20')
plt.title('Salinas  Random Forest Classification')
plt.colorbar()
plt.axis('off')
plt.show()
```
